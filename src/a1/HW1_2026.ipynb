After your yearly checkup, the doctor has bad news and good news. The
bad news is that you tested positive for a very serious cancer and
that the test is 99.5% accurate i.e. the probability of testing
positive given you have the disease is 0.995. The probability of
testing negative if you don‚Äôt have the disease is the same (also 0.995). The good news is that it is a rare condition affecting only 1 in 500 people. **What is the probability you actually have the disease?** 

After doing all your calculations you realize that there was a misprint on the test, and the accuracy was actually only 95% (for both testing postive given that you have the disease and for testing negative given that you do not have the disease). **How will this change your probability of having the disease?**

Show all calculations and the final result. Remember to use LaTeX math-mode to format mathematical expressions and equations.


#### üìù Your answer here: 
##### A:
Let the prevelance $P(D)$ = $\frac{1}{500} \implies P(\neg D) = 1 - P(D) = \frac{499}{500}$. The true positive and negative rate is $P(+|D) = 0.995 = P(-|\neg D)$ and thus the false positive rate is $P(+ | \neg D) = 1 - P(-|\neg D) = 1 - 0.995 = 0.005$. From here bayes formula states:

\begin{equation}
P(D | +) = \frac{P(+ | D)P(D)}{P(+|D)P(D) + P(+|\neg D )P(\neg D)} = \frac{0.995 \cdot 0.002}{0.995 \cdot 0.002 + 0.005 \cdot 0.998} = 0.2851...
\end{equation}
Meaning despite a positive test the chance of still having the disease is just $28.51 \%$

##### B:
Now the true positive/negative is $0.95$ meaning the false positive probability is $P(+ | \neg D) = 1 - P(-|\neg D) = 1 - 0.95 = 0.05$. The prevelance remains the same and thus:

\begin{equation}
P(D | +) = \frac{P(+ | D)P(D)}{P(+|D)P(D) + P(+|\neg D )P(\neg D)} = \frac{0.95 \cdot 0.002}{0.95 \cdot 0.002 + 0.05 \cdot 0.998} = 0.03668...
\end{equation}
Meaning with $95 \%$ true positive rate we still only have roughly $3.7\%$ chance of having the disease given a positive result.

## Problem 4: Spherical Gaussian estimation [4 points]

Consider a dataset $X$ consisting of i.i.d. observations
generated from a spherical Gaussian distribution $N(\mu, \sigma^2I)$, where $\mu \in \mathbb{R}^p$, $I$ 
is the $p \times p $ identity matrix, and $\sigma^2$ is a scalar.

Write the mathematical expression for the Maximum Likelihood Estimator (MLE) for $\mu$ and $\sigma$ in above setup.

For more information about the spherical Gaussian distribution, see https://en.wikipedia.org/wiki/Multivariate_normal_distribution.
For more information about the identity matrix see: https://en.wikipedia.org/wiki/Identity_matrix

#### üìù Your answer here: 
From the wiki we find the pdf of a single observation is given by
\begin{equation}
f(\vec{x}_i) = (2\pi)^{-\frac{p}{2}}det(\mathbf{\Sigma})^{-\frac{1}{2}} e^{-\frac{1}{2}(\vec{x}_i-\vec{\mu})^T\mathbf{\Sigma}^{-1}(\vec{x}_i - \vec{\mu})} = (2\pi)^{-\frac{p}{2}}det(\sigma^2 \mathbf{I})^{-\frac{1}{2}} e^{-\frac{1}{2}(\vec{x}_i-\vec{\mu})^T(\sigma^2 \mathbf{I})^{-1}(\vec{x}_i - \vec{\mu})}
\end{equation}

Meaning the likelihood for the entire dataset is (assuming $n$ i.i.d. observations)

\begin{equation}
\pi(X | \sigma^2, \vec{\mu}) = \Pi_{i=1}^n f(\vec{x}_i) = \Pi_{i=1}^n(2\pi)^{-\frac{p}{2}}det(\sigma^2 \mathbf{I})^{-\frac{1}{2}} e^{-\frac{1}{2}(\vec{x}_i-\vec{\mu})^T(\sigma^2 \mathbf{I})^{-1}(\vec{x}_i - \vec{\mu})}.
\end{equation}
From here we determine the log likelihood in order to simplify future derivatives (from here $\sum$ denotes "sum")
\begin{equation}
ln(\pi(X | \sigma^2, \vec{\mu})) = \sum_{i=1}^n ln[f(\vec{x}_i)] = \sum_{i=1}^n[ -\frac{p}{2}ln(2\pi)-\frac{1}{2}ln(det(\sigma^2 \mathbf{I}))-\frac{1}{2}(\vec{x}_i-\vec{\mu})^T(\sigma^2 \mathbf{I})^{-1}(\vec{x}_i - \vec{\mu})].
\end{equation}

From here we need to determine $det[\sigma^2 \mathbf{I}]$ and $(\sigma^2 \mathbf{I})^{-1}$. The determinant is calculated by multiplying diagonals, since we have the identity matrix of dimension $p \times p$ we end up multiplying $\sigma^2$, $p$ times, all other contributions are $0$. Thus, $det[\sigma^2 I]$ = $(\sigma^2)^p$. The inverse is also quite simple since we are working with the identity matrix, $(\sigma^2 I) (\frac{1}{\sigma^2}I) = I$ and thus $(\sigma^2 I)^{-1} = (\frac{1}{\sigma^2}I)$, $\sigma^2>0$. Inserting this we get the following expression for the log likelihood.

\begin{equation}
ln(\pi(X | \sigma^2, \vec{\mu})) = -\frac{np}{2}ln(2\pi) - \frac{np}{2} ln(\sigma^2)-\frac{1}{2\sigma^2}\sum_{i=1}^n[(\vec{x}_i-\vec{\mu})^TI(\vec{x}_i - \vec{\mu})]
 = -\frac{np}{2}ln(2\pi) - \frac{np}{2} ln(\sigma^2)-\frac{1}{2\sigma^2}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}||^2
\end{equation}
In the last step we used the fact that a vector $\vec{v}^T$ with dimensions $1 \times p$ multiplied by the identity matrix $p \times p$ is the same vector $\vec{v}^T$. We then use the definition of the euclidian norm. From here we want to find $\vec{\mu}, \sigma^2$ maximizing $ln(\pi(X | \vec{\mu}, \sigma^2))$. The only term including $\mu$ is the last one. We also notice that if we minimize the term we maximize the log likelihood. Thus we start by deriving an expression for its derivative. We call the term $g(\vec{\mu}) = \sum_{i=1}^n||\vec{x}_i - \vec{\mu}||^2$,

\begin{equation}
\frac{\partial g(\mu)}{\partial \mu_k} = \frac{\partial}{\partial \mu_k}[\sum_{i=1}^n||\vec{x}_i - \vec{\mu}||^2] = \sum_{i=1}^n\frac{\partial}{\partial \mu_k}||\vec{x}_i - \vec{\mu}||^2 = \sum_{i=1}^n\frac{\partial}{\partial \mu_k}(x_{ik} - \mu_k)^2 = \sum_{i=1}^n 2 (x_{ik} - \mu_k)(-1) = 2[\sum_{i=1}^n (\mu_k - x_{ik})] = 2[n\mu_k - \sum_{i=1}^n (x_{ik})] = 0.
\end{equation}

\begin{equation}
\implies \mu_{k} = \frac{1}{n} \sum_{i=1}^n (x_{ik})
\end{equation}

Thus, the full vector notation is $\vec{\mu}_{MLE} = \frac{1}{n} \sum_{i=1}^n (\vec{x}_{i})$, we also note that this is the average over observations. From here we reinsert the MLE for $\mu$ and instead derivate w.r.t. $\sigma$,

\begin{equation}
\frac{\partial ln(\pi(X | \sigma^2, \vec{\mu}_{MLE}))}{\partial \sigma} = \frac{\partial}{\partial \sigma}[-\frac{np}{2}ln(2\pi) - \frac{np}{2} ln(\sigma^2)-\frac{1}{2\sigma^2}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}_{MLE}||^2] = \frac{\partial}{\partial \sigma}[- \frac{np}{2} ln(\sigma^2)] + \frac{\partial}{\partial \sigma}[-\frac{1}{2\sigma^2}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}_{MLE}||^2] = \frac{-np}{\sigma} + \frac{1}{\sigma^3}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}_{MLE}||^2 = 0
\end{equation}

\begin{equation}
\implies \sigma^2 = \frac{1}{np}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}_{MLE}||^2.
\end{equation}

Thus, finally
\begin{equation}
\sigma_{MLE} = \sqrt{\frac{1}{np}\sum_{i=1}^n||\vec{x}_i - \vec{\mu}_{MLE}||^2}.
\end{equation}
